```python
#splitting the parameters and target variable
X = data.drop(columns='y',axis=1)
Y = data['y']
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

# Pipeline: önce veriyi scale et, sonra Lasso uygula
pipeline = make_pipeline(StandardScaler(), Lasso())

param_grid = {
    'lasso__alpha': [0.1, 1.0, 10.0]  # dikkat: 'lasso__' diye yazılıyor pipeline için!
}

grid = GridSearchCV(pipeline, param_grid, cv=10)
grid.fit(X_train, Y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
import pandas as pd
res = pd.pivot_table(pd.DataFrame(grid.cv_results_), values = 'mean_test_score', index = 'param_lasso__alpha')
```
