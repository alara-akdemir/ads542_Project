### Hyperparameter Tuning

```python
param_grid = {'alpha':np.logspace(-3,0,13)}
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

# Pipeline:
pipeline = make_pipeline(StandardScaler(), Lasso())

param_grid = {
    'lasso__alpha': [0.01, 0.1, 1.0, 10.0, 100.0],
    'lasso__fit_intercept': [True, False],
    'lasso__selection': ['cyclic', 'random'],
    'lasso__max_iter': [1000, 5000, 10000],
    'lasso__tol': [1e-4, 1e-3, 1e-2]
}

grid = GridSearchCV(pipeline, param_grid, cv=10)
grid.fit(X_train, Y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
import pandas as pd
res = pd.pivot_table(pd.DataFrame(grid.cv_results_),values=['mean_test_score', 'std_test_score', 'mean_fit_time', 'mean_score_time', 'rank_test_score'],   index=['param_lasso__alpha', 'param_lasso__fit_intercept', 'param_lasso__selection', 'param_lasso__max_iter', 'param_lasso__tol'])
```

```python
res
```

### Evaluation

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
# Create best_model pipeline with best parameters
best_model = make_pipeline(
    StandardScaler(),
    Lasso(
        alpha=0.01,
        fit_intercept=True,
        max_iter=1000,
        selection='random',
        tol=0.01
    )
)
```

```python
# Fit the best_model on the training data
best_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
# Predict on X_test
y_pred = best_model.predict(X_test)

# Converting continuous predictions to binary (assuming threshold=0.5)
y_pred_class = (y_pred >= 0.5).astype(int)

# Then print classification report
print(classification_report(Y_test, y_pred_class))
```

```python
import numpy as np

# Input data
input_data = (0, 30, 1, 1, 2, 0, 2, 0, 0, 6, 0, 2, 999, 0, 1, -1.8, 92.893, -46.2, 1.313, 0)

# Convert to numpy array
input_data_array = np.asarray(input_data)

# Reshape for single instance prediction
input_data_reshaped = input_data_array.reshape(1, -1)

# Make prediction using best_model
prediction = best_model.predict(input_data_reshaped)

# Since Lasso is regression → output is continuous → classify using threshold
predicted_class = int(prediction >= 0.5)

print('Credit Approval; 0 for rejection, 1 for acceptance :', predicted_class)
```
